## Background

In recent years, Singapore has experienced a surge in fusion cuisine, blending international flavors with local delicacies. While this culinary trend has gained popularity, it raises concerns about preserving the authenticity and traditional essence of local dishes.

Amidst this dynamic gastronomic landscape, there's a delicate balance to be struck. Fusion cuisine caters to those who appreciate international twists, drawing attention to Singapore's culinary scene globally. However, this shift also prompts reflection on the preservation of authentic local experiences.

The ongoing dialogue between fusion and tradition reflects diverse culinary preferences. While some gravitate towards the exciting blend of international flavors, others seek authentic tasting food true to the essence of their cultural roots. Navigating this intricate interplay between innovation and tradition becomes crucial in defining Singapore's culinary identity.

```python
def plot_word_cloud(df, restaurant_label=None):
    
    df = df[df['restaurant_label']!=0]
    # Filter the dataframe if a restaurant_label is provided
    if restaurant_label is not None:
        df = df[df['restaurant_label'] == restaurant_label]

    # Assuming 'review_text' is the column containing cleaned and lemmatized text
    all_reviews_text = ' '.join(df['review_text'])

    # Define custom stopwords
    custom_stopwords = set(['restaurant', 'service', 'well', 'place', 'serve', 'singapore', 'go', 'order', 'good', 'dish','nice','indian', 'one', 'try', 'also', 'food'])

    # Create a word cloud with custom stopwords
    wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=custom_stopwords).generate(all_reviews_text)

    # Plot the WordCloud image                        
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    
    if restaurant_label is not None:
        plt.title(f'Top Words for Restaurant {restaurant_label}')
    else:
        plt.title('Top Words Across All Reviews')
    
    plt.show()
```

### Top nouns and adjectives

```python
# function that plot and can filter by restaurant, the top nouns and adjectives from reviews, with restaurant 0
```

```python
def plot_top_nouns(df, top_n=10, restaurant_label=None):
    # Filter the dataframe if a restaurant_label is provided
    if restaurant_label is not None:
        df = df[df['restaurant_label'] == restaurant_label]
        restaurant = df['restaurant'].unique()[0]

    # Assuming 'review_text' is the column containing cleaned and lemmatized text
    all_reviews_text = ' '.join(df['review_text'])

    # Tokenize and lemmatize the text
    tokens = word_tokenize(all_reviews_text.lower())
    lemmatized_words = [word for word in tokens if word.isalpha()]

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    meaningful_words = [word for word in lemmatized_words if word not in stop_words]

    # Perform part-of-speech tagging
    pos_tags = pos_tag(meaningful_words)

    # Filter for nouns (NN) or adjectives (JJ)
    selected_tags = [word for word, tag in pos_tags if tag in ['NN', 'NNS', 'JJ', 'JJR', 'JJS']]

    # Count occurrences
    word_counts = Counter(selected_tags)

    # Extract top n words and their counts
    top_words = [word for word, count in word_counts.most_common(top_n)]
    top_counts = [count for word, count in word_counts.most_common(top_n)]

    # Plot the top nouns using Seaborn barplot
    plt.figure(figsize=(12, 6))
    sns.barplot(x=top_words, y=top_counts, palette='viridis')
    
    if restaurant_label is not None:
        plt.title(f'Top {top_n} Nouns/Adjectives for Restaurant {restaurant_label}, {restaurant}')
    else:
        plt.title(f'Top {top_n} Nouns/Adjectives Across All Reviews')
        
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
plot_top_nouns(data, top_n=30)
```

Authentic is in the top 20 words in the reviews, coupled with the quality and taste of the dishes as well as the experience at the restaurants.

```python
# A function to obtain an n-gram from the reviews
# the review_text column is already cleaned and the words are all in lower case

def generate_ngrams(df, label, n=2):
    
    # Get the text based on the provided restaurant_label or for all reviews
    filtered_df = df[df['restaurant_label'] == label]

    # Tokenize and remove stopwords from the review_text column
    stop_words = set(stopwords.words('english'))
    filtered_df['review_tokens'] = filtered_df['review_text'].apply(lambda x: [word.lower() for word in word_tokenize(x) if word.isalnum() and word.lower() not in stop_words])

    # Generate n-grams
    filtered_df['ngrams'] = filtered_df['review_tokens'].apply(lambda x: list(ngrams(x, n)))

    return filtered_df[['restaurant_label', 'review_text', 'ngrams']]
```

```python
# Function to get the top n-grams and their frequencies
def get_top_ngrams(ngram_list, top_n=10):
    # Flatten the list of n-grams
    flattened_ngrams = [item for sublist in ngram_list for item in sublist]

    # Count the occurrences of each n-gram
    ngram_counts = Counter(flattened_ngrams)

    # Sort n-grams by frequency in descending order
    sorted_ngrams = sorted(ngram_counts.items(), key=lambda x: x[1], reverse=True)[:top_n]

    return sorted_ngrams
```

```python
# A function to plot the top ngrams

def plot_top_items(top_items, ngram_size=2, title='Top Items', xlabel='Frequency', ylabel='Item', figsize=(10, 6), color='skyblue'):
    # Extract items and frequencies
    items, frequencies = zip(*[(' '.join(item), freq) for item, freq in top_items])

    # Adjust title based on n-gram size
    title += f' ({ngram_size}-grams)'

    # Plot a horizontal bar chart
    plt.figure(figsize=figsize)
    plt.barh(items, frequencies, color=color)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

## Sentiment Analysis

```python
def sentiment_scores(sentence):

	# Create a SentimentIntensityAnalyzer object.
	sia = SentimentIntensityAnalyzer()

	# polarity_scores method of SentimentIntensityAnalyzer
	# object gives a sentiment dictionary.
	# which contains pos, neg, neu, and compound scores.
	sentiment_dict = sia.polarity_scores(sentence)
	return sentiment_dict
```

```python
sentiment_df = data['review_text'].apply(lambda x: pd.Series(sentiment_scores(x)))

# Concatenate the sentiment scores DataFrame with the original DataFrame
result_df = pd.concat([data, sentiment_df], axis=1)
```

```python
def plot_sentiment_distribution(df, restaurant_label=None):
    
    if restaurant_label is not None:
        df = df[df['restaurant_label'] == restaurant_label]

    sentiment_columns = ['neg', 'neu', 'pos']

    # Group by restaurant_label and calculate mean sentiment scores
    grouped_df = df.groupby('restaurant_label')[sentiment_columns].mean()
    
    # Plot a stacked bar chart
    grouped_df.plot(kind='bar', stacked=True, colormap='viridis', figsize=(14, 8))
    
    if restaurant_label is not None:
        plt.title(f'Sentiment Distribution for Restaurant {restaurant_label}')
    else:
        plt.title('Sentiment Distribution Across All Reviews')
    
    
    plt.xlabel('Restaurant Label')
    plt.ylabel('Average Sentiment Score')
    plt.legend(['Negative', 'Neutral', 'Positive'], loc='upper right')
    plt.tight_layout()
    plt.show()
```

```python
plot_sentiment_distribution(result_df)
```

This chart shows the proportion of negative, neutral and positive sentiments of each restaurant. Most of them have more reviews with neutral sentiments as compared to positive sentiments.

```python
def sentiment_dist_hist(df,sent):
    
    color = {'neg': 'red', 'neu': 'blue', 'pos':'green', 'compound':'orange'}
    
    sentiment = df[sent]
    
    plt.hist(sentiment, bins=10, color=color[sent], alpha=0.7)
    
    plt.axvline(sentiment.mean(), color='maroon', linestyle='dashed', linewidth=2, label='Mean')

    plt.title(f'Distribution of {sent} Sentiments')
    plt.xlabel(f'{sent} Sentiment Score')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# A function that filters out the words of of reviews where the score of sentiment1 is greater than the score of sentiment2.

def filtered_words(df,sentiment1, sentiment2, num_words):

    filtered_reviews = df[df[sentiment1] > df[sentiment2]]

    # Tokenize and preprocess the text
    stop_words = set(stopwords.words('english'))
    filtered_reviews['tokenized_text'] = filtered_reviews['review_text'].apply(lambda x: word_tokenize(x.lower()))
    filtered_reviews['filtered_text'] = filtered_reviews['tokenized_text'].apply(lambda x: [word for word in x if word.isalpha() and word not in stop_words])

    # Count the occurrences of words
    all_words = [word for review in filtered_reviews['filtered_text'] for word in review]
    word_counts = Counter(all_words)

    # Display or visualize the most common words
    common_words_df = pd.DataFrame(word_counts.most_common(num_words), columns=['Word', 'Count'])

    # Visualize the word frequencies
    plt.figure(figsize=(14, 6))
    plt.bar(common_words_df['Word'], common_words_df['Count'])
    plt.xlabel('Word')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Frequency')
    plt.title(f'Top Words in Reviews with {sentiment1} > {sentiment2} sentiment')
    plt.show()
```
